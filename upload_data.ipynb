```python
from pathlib import Path
from datetime import date, timedelta

import httpx

folder = Path.cwd() / "data_CO"
folder.mkdir(exist_ok=True, parents=True)

start_date = date(2023, 1, 1)
url = "https://luftdaten.umweltbundesamt.de/api-proxy/measures/csv"
for i in range(365):
    current_date = start_date + timedelta(days=i)
    file = folder / f"{current_date.isoformat()}.csv"

    if file.exists():
        continue

    
    params = {
        "date_from": current_date.isoformat(),
        "time_from": "1",
        "date_to": current_date.isoformat(),
        "time_to": "24",
        "data[0][co]": "2",  # 2 = CO
        "data[0][sc]": "4",  # 4 = 8SMW; 2 = 1SMW
        "lang": "de"
    }

    response = httpx.get(url, params=params, timeout=60)
    
    file.write_text(response.content.decode("utf-8"))
    from time import sleep

    #sleep(30)
```
