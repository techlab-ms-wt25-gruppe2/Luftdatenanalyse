```python
# df1 und df2 über eine gemeinsame Spalte 'ID' zusammenführen
df_combined = pd.merge(df1, df2, left_on='NUTS Name', right_on='Stationsname')
print(df_combined.head())
```

```python
# axis=1 steht für Spalten (axis=0 wären Zeilen)
# how='all' bedeutet: Lösche nur, wenn ALLES leer ist
df_combined = df_combined.dropna(axis=1, how='all')
```

```python
print(df_combined.info())
```

```python
print(df_combined.describe())
```

```python
# Löscht Zeilen, die KOMPLETT identisch sind
df_combined = df_combined.drop_duplicates()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Nur numerische Spalten auswählen (Text kann nicht korrelieren)
df_num = df_combined.select_dtypes(include=['number'])

# 2. Korrelationsmatrix berechnen
correlation_matrix = df_num.corr()

# 3. Als Heatmap zeichnen
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Korrelation der Spalten")
plt.show()
```

```python
import numpy as np

# 1. Korrelationsmatrix berechnen (Absolutwerte, da -0.99 auch stark ist)
corr_matrix = df_combined.select_dtypes(include=['number']).corr().abs()

# 2. Nur das obere Dreieck der Matrix betrachten
# (Sonst findest du A zu B und B zu A doppelt)
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# 3. Spalten finden, die eine Korrelation > 0.95 haben
zu_loeschen = [column for column in upper.columns if any(upper[column] > 0.95)]

print(f"Folgende Spalten sind redundant und können gelöscht werden: {zu_loeschen}")

# 4. Optional: Direkt löschen
df_bereinigt = df_combined.drop(columns=zu_loeschen)
```
